```python
import pandas as pd
UnfilteredDF = pd.read_csv('Unfiltered Data.csv')
FilteredDF = pd.read_csv('Filtered Data.csv')
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(12, 6), dpi=80)
Plot = plt.bar(UnfilteredDF['Chromosome'], UnfilteredDF['Count'], color='red', edgecolor = 'black')

#Set every other bar to be blue
for i in range(0,len(UnfilteredDF['Chromosome'].tolist())):
    if i%2 !=0:
        Plot[i].set_color('blue')
        Plot[i].set_edgecolor('black')
        
plt.title('Indel  Count per Chromosome (Unfiltered)', fontsize=14)
plt.xlabel('Chromosome', fontsize=14)
plt.ylabel('Number of Indels  $(x10^{6})$', fontsize=14)
plt.grid(False)
plt.savefig("Unfiltered.png",dpi=300)
plt.show()
```

```python
figure(figsize=(12, 6), dpi=80)
Plot = plt.bar(FilteredDF['Chromosome'], FilteredDF['Count'], color='red',edgecolor='black')

#Set every other bar to be blue with a black border
for i in range(0,len(FilteredDF['Chromosome'].tolist())):
    if i%2 !=0:
        Plot[i].set_color('blue')
        Plot[i].set_edgecolor('black')
        
plt.title('Indel Count per Chromosome (Filtered)', fontsize=14)
plt.xlabel('Chromosome', fontsize=14)
plt.ylabel('Number of Indels', fontsize=14)
plt.grid(False)
plt.savefig("Filtered.png",dpi=300)
plt.show()
```

```python
import numpy as np
figure(figsize=(12, 6), dpi=80)
n=4
r = np.arange(len(FilteredDF['Chromosome'].tolist()))
width = 0.35
  
  
plt.bar(r, UnfilteredDF['Count'], color = 'b',
        width = width, edgecolor = 'black',
        label='Unfiltered')
plt.bar(r + width, FilteredDF['Count'], color = 'r',
        width = width, edgecolor = 'black',
        label='Filtered')
  
plt.xlabel("Chromosome",fontsize=14)
plt.ylabel('Number of Indels  $(x10^{6})$', fontsize=14)
plt.title("Indel Count per Chromosome")
  
# plt.grid(linestyle='--')
plt.xticks(r + width/2,FilteredDF['Chromosome'].tolist())
plt.legend()
plt.savefig("Combined.png",dpi=300)
plt.show()
```
