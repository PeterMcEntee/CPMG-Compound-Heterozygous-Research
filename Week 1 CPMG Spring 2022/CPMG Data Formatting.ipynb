```python
# importing csv module
import csv
  
# csv file name
filename = "TestData.csv"
  
# initializing the titles and rows list
fields = []
rows = []
  
# reading csv file
with open(filename, 'r') as csvfile:
    # creating a csv reader object
    csvreader = csv.reader(csvfile)
      
    # extracting field names through first row
    fields = next(csvreader)
  
    # extracting each data row one by one
    for row in csvreader:
        rows.append(row)
```

```python
#Extracting CSV Headers
NewFields = [i.split('\t') for i in fields][0]
NewFields
```

```python
#Creating Rows for pandas DataFrame
NewRows = []
for row in rows:
    for i in row:
        NewRows.append(i.split('\t'))
```

```python
#Load CSV data in as DataFrame
import pandas as pd
df = pd.DataFrame(NewRows,columns = NewFields)
df
```

```python
ChromList = df['Chr-Pos'].tolist()
```

```python
ChromePosTotal = []
for i in range(0,len(ChromList)):
    Chromosome,Pos = ChromList[i].split('-')
    ChromePosTotal.append((Chromosome,int(Pos)))
```

```python
Chrom_Pos_Dict = {}
for i in ChromePosTotal:
    if i[0] not in Chrom_Pos_Dict:
        Chrom_Pos_Dict[i[0]] = []
    else:
        Chrom_Pos_Dict[i[0]].append(i[1])
        
```

```python
for key, value in Chrom_Pos_Dict.items():
    print(key,len(value))
```

```python
ChromCount = [[key,len(value)] for key, value in Chrom_Pos_Dict.items()]
```

```python
ChromDF = pd.DataFrame(ChromCount,columns=['Chromosome','Count'],dtype='int')
ChromDF
```

```python
Chromosomes = ChromDF['Chromosome'].tolist()
for i in range(0,len(Chromosomes)):
    try:
        Chromosomes[i] = int(Chromosomes[i])
    except:
        ValueError

ChromDF['Chromosome'] = Chromosomes
```

```python
Chromosomes
```

```python
SexChroms = ChromDF.iloc[[22,23]]
ChromDF.drop([22,23], axis=0, inplace=True)
ChromDF = ChromDF.sort_values(by="Chromosome")
ChromDF = ChromDF.append(SexChroms)
```

```python
ChromDF
```

```python
## Applying Filters to Original CSV:
MultiSample,MultiAllele,NearbyIndel = df['MultiSampleCount'].tolist(),df['MultiAlleleCount'].tolist(),df['NearbyIndelCount25bp'].tolist()
for i in range(0,len(MultiSample)):
    try:
        MultiSample[i] = int(MultiSample[i])
        MultiAllele[i] = int(MultiAllele[i])
        NearbyIndel[i] = int(NearbyIndel[i])
    except:
        ValueError

df['MultiSampleCount'] = MultiSample
df['MultiAlleleCount'] = MultiAllele
df['NearbyIndelCount25bp'] = NearbyIndel
```

```python
filtered_df = df[(df["MultiSampleCount"] >= 4) | (df["MultiAlleleCount"] >= 4) | (df["NearbyIndelCount25bp"] >= 4)]
filtered_df
```

```python
filtered_chrom = filtered_df['Chr-Pos'].tolist()
Filtered_ChromPosTotal = []
for i in range(0,len(filtered_chrom)):
    Chromosome,Pos = filtered_chrom[i].split('-')
    Filtered_ChromPosTotal.append((Chromosome,int(Pos)))
```

```python
Filtered_Chrom_Pos_Dict = {}
for i in Filtered_ChromPosTotal:
    if i[0] not in Filtered_Chrom_Pos_Dict:
        Filtered_Chrom_Pos_Dict[i[0]] = []
    else:
        Filtered_Chrom_Pos_Dict[i[0]].append(i[1])
        
```

```python
Filtered_ChromCount = [[key,len(value)] for key, value in Filtered_Chrom_Pos_Dict.items()]
```

```python
Filtered_ChromDF = pd.DataFrame(Filtered_ChromCount,columns=['Chromosome','Count'],dtype='int')
Chromosomes = Filtered_ChromDF['Chromosome'].tolist()
for i in range(0,len(Chromosomes)):
    try:
        Chromosomes[i] = int(Chromosomes[i])
    except:
        ValueError

Filtered_ChromDF['Chromosome'] = Chromosomes
```

```python
SexChroms = Filtered_ChromDF.iloc[[22,23]]
Filtered_ChromDF.drop([22,23], axis=0, inplace=True)
Filtered_ChromDF = Filtered_ChromDF.sort_values(by="Chromosome")
Filtered_ChromDF = Filtered_ChromDF.append(SexChroms)
Filtered_ChromDF
```

```python
Filtered_ChromDF.to_csv('Filtered Data.csv',index=False)
ChromDF.to_csv('Unfiltered Dat a.csv',index=False)
```

```python
#Storing DataFrames at HTML Tables:
Unfiltered_html = ChromDF.to_html(index=False)
Filtered_html = Filtered_ChromDF.to_html(index=False)
```

```python
 # write html to file
text_file = open("UnfilteredTable.html", "w",)
text_file.write(Unfiltered_html)
text_file.close()
```
